```python
df = pd.read_csv('DrugData.csv')
```

```python
data = df.loc[:,['품목명', '큰제품이미지']]
data.head()
```

```python
def get_download(url, fname, directory):
    try:
        os.chdir(directory)
        request.urlretrieve(url, fname)
        print(fname,"다운로드 완료")
    except HTTPError as e:
        print('error')
        return

path = "C:\\woodystudio\\OOP7\\drug_image"
for i in range (374, len(data)):
    filename = str(data.loc[i]['품목명']) + ".jpg"
    url = data.loc[i]['큰제품이미지']
    get_download(url, filename, path)
```
